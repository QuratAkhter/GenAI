```python
import pandas as pd
```

```python
df = pd.read_csv('DisneylandReviews.csv',encoding="latin-1")
```

```python
df = df['Review_Text']
```

```python
# Lower casing

df = df.apply(lambda x: " ".join(x.lower() for x in x.split()))
```

```python
df.head()
```

```python
# Removing HTML Tags if any
import re
def remove_html_tags(text):
    pattern = re.compile('<.*?>')
    return pattern.sub(r'', text)

df = df.apply(lambda x: remove_html_tags(x))
```

```python
# Removing URL if any

def remove_url(text):
    pattern = re.compile(r'https?://\S+|www\.\S+')
    return pattern.sub(r'',text)

df = df.apply(lambda x: remove_url(x))
```

```python
# chat word conversion
chat_words = {
    'AFAIK':'As Far As I Know',
    'AFK':'Away From Keyboard',
    'ASAP':'As soon as possible',
    'ATK':'At The Keyboard',
    'ATM':'At The Moment',
    'A3':'Anytime, Anywhere, Anyhow',
    'BAK':'Back At Keyboard',
    'BBL':'Be Back Later',
    'BBS':'Be Back Soon',
    'BFN':'Bye For Now',
    'B4N':'Bye For Now',
    'BRB':'Be Right Back',
    'BRT':'Be Right There',
    'BTW':'By The Way',
    'B4':'Before',
    'B4N':'Bye For Now',
    'CU':'See You',
    'CUL8R':'See You Later',
    'CYA':'See You',
    'FAQ':'Frequently Asked Questions',
    'FC':'Fingers Crossed',
    'FWIW':'For What It\'s Worth',
    'FYI':'For Your Information',
    'GAL':'Get A Life',
    'GG':'Good Game',
    'GN':'Good Night',
    'GMTA':'Great Minds Think Alike',
    'GR8':'Great!',
    'G9':'Genius',
    'IC':'I See',
    'ICQ':'I Seek you',
    'ILU':'I Love You',
    'IMHO':'In My Honest/Humble Opinion',
    'IMO':'In My Opinion',
    'IOW':'In Other Words',
    'IRL':'In Real Life',
    'KISS':'Keep It Simple, Stupid',
    'LDR':'Long Distance Relationship'
}

def chat_conversion(text):
    for word in text.split():
      if word in chat_words:
        text = text.replace(word,chat_words[word])
    return text

df = df.apply(lambda x: chat_conversion(x))
```

```python
df.head()
```

```python
# Incorrect word handling

from textblob import TextBlob
df = df.apply(lambda x: str(TextBlob(x).correct()))
```

```python
df.head()
```

```python
# stop words removal

import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```python
def remove_stopwords(text):
    for word in text.split():
      if word in stopwords.words('english'):
        text = text.replace(word,'')
        return text

df = df.apply(lambda x: remove_stopwords(x))
```

```python
# Emoji Handling
import emoji

def handle_emoji(text):
  return emoji.demojize(text)

df =df.apply(lambda x: handle_emoji(x))
```

```python
# Tokenize

from nltk.tokenize import word_tokenize
import nltk
nltk.download('punkt')
nltk.download('punkt_tab')

df = df.apply(lambda x: word_tokenize(x))
```

```python
# Lemmatization

from nltk.stem import WordNetLemmatizer
from nltk.corpus import wordnet
nltk.download('wordnet')
nltk.download('averaged_perceptron_tagger_eng')
lemmatizer = WordNetLemmatizer()
```

```python
def get_wordnet_pos(tag):
    if tag.startswith('J'):
        return wordnet.ADJ
    elif tag.startswith('V'):
        return wordnet.VERB
    elif tag.startswith('N'):
        return wordnet.NOUN
    elif tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN

text = "Hello dearest I am too goods"
word_pos_tags = nltk.pos_tag(word_tokenize(text))
print(word_pos_tags)
```

```python
def lemmatize_text(text):
  # tokens = word_tokenize(text)
  pos_tags = nltk.pos_tag(text)
  return " ".join(
        lemmatizer.lemmatize(word, get_wordnet_pos(tag))
        for word, tag in pos_tags
    )
```

```python
df = df.apply(lambda x: lemmatize_text(x))
```
