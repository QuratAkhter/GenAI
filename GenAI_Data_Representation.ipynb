# Bag of Words (BoW)

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer()
```

```python
bow = cv.fit_transform(df['text'])
```

```python
cv.vocabulary_
```

```python
bow.toarray()
```

```python
cv.transform(['Annie watch Tom and Jerry']).toarray()
```

# n-Gram

```python
cv2 = CountVectorizer(ngram_range=(2,2))
```

```python
ngram = cv2.fit_transform(df['text'])
```

```python
cv2.vocabulary_
```

```python
ngram.toarray()
```

```python
cv2.transform(['Annie write comment']).toarray()
```

# Word2Vec

It is a deep learning based method that understand the semantic information

```python
import gensim
import os
```

```python
from nltk import sent_tokenize
from gensim.utils import simple_preprocess
import nltk
nltk.download('punkt')
nltk.download('punkt_tab')
```

```python
story = []

for filename in os.listdir('data'):
    if filename == '.ipynb_checkpoints':
      pass
    f = open(os.path.join('data',filename))
    corpus = f.read()
    raw_sent = sent_tokenize(corpus)
    for sent in raw_sent:
      story.append(simple_preprocess(sent))
```

```python
story
```

```python
len(story)
```

```python
story[0]
```

```python
model = gensim.models.Word2Vec(
    window=10,
    min_count=2
)
```

```python
model.build_vocab(story)
```

```python
model.train(story, total_examples=model.corpus_count, epochs=model.epochs)
```

```python
model.wv.most_similar('tom')
```

```python
model.wv.most_similar('dothraki')
```

```python
### Visualizing the vectors

from sklearn.decomposition import PCA
```

```python
vec = model.wv.get_normed_vectors()
```

```python
vec.shape
```

```python
pca = PCA(n_components=3)
```

```python
X = pca.fit_transform(vec)
```
